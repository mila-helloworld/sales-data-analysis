# Sales Data Analysis

This project explores a mock sales dataset generated with Mockaroo.  
Weâ€™ll perform exploratory data analysis (EDA) and visualize sales trends  
by region, category, and product.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sales_data.csv")

print(df.head())
print(df.info())
print(df.describe())

sales_region = df.groupby("region")["revenue"].sum()
sales_region.plot(kind="bar", title="Total Sales by Region")
plt.show()

sales_category = df.groupby("category")["revenue"].sum()
sales_category.plot(kind="bar", title="Total Sales by Category")
plt.show()

sales_product = df.groupby("product")["revenue"].sum().sort_values(ascending=False)
sales_product.plot(kind="bar", title="Total Sales by Product")
plt.show()
```

## Missing Values

## Sales by Region
We aggregate total revenue by region to see where sales are strongest.

```python
# Total revenue by region
sales_region = df.groupby("region")["revenue"].sum()
sales_region.plot(kind="bar", title="Total Sales by Region")
plt.show()
```

## Sales by Category
We compare revenue contributions across product categories.

```python
# Total revenue by category
sales_category = df.groupby("category")["revenue"].sum()
sales_category.plot(kind="bar", title="Total Sales by Category")
plt.show()
```

## Top Products by Revenue
Identify the top 10 products driving the highest revenue.

```python
# Top 10 products by revenue
top_products = df.groupby("product")["revenue"].sum().sort_values(ascending=False).head(10)
top_products.plot(kind="bar", title="Top 10 Products by Revenue")
plt.show()
```
